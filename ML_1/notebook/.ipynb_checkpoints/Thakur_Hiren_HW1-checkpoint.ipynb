### (a) Download Data

Package imports

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import DistanceMetric
import warnings
warnings.filterwarnings('ignore')
```

Get the Vertebral Column Data Set

```python
filepath = '..\\data\\column_2C.dat'


name_columns = [
    'pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle', 
    'sacral_slope', 'pelvic_radius', 'degree_spondylolisthesis', 'class']


data = pd.read_csv(filepath, sep="\\s+", header=None, names=name_columns)
```

```python
data
```

#### i. Scatterplots

```python
data['class'] = data['class'].replace({'NO':0, 'AB':1})


scatter_plot = sns.pairplot(data,diag_kind = 'kde', hue = 'class', palette=['g','y'])

plt.show()
```

#### ii. Boxplots

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 20))

for ax, columns in zip(axes.flatten(), name_columns):
    
    sns.boxplot(x='class', y=columns, data=data, ax=ax, hue= 'class' , palette=['r','b'] )
   

plt.show()
```

#### iii. Split Data Set

```python
zero_data = data[data['class']==0]
one_data = data[data['class']==1]

train_0, test_0 = train_test_split(zero_data, train_size=70, random_state=42, shuffle= False)
train_1, test_1 = train_test_split(one_data, train_size=140, random_state=42, shuffle= False)

trained_set = pd.concat([train_0, train_1])
tested_set = pd.concat([test_0, test_1])

print(trained_set.shape)
print(tested_set.shape)
```

### (c) Classification

#### i. Euclidean Metric

```python
X_train = trained_set.drop('class', axis=1)  
y_train = trained_set['class']               
X_test = tested_set.drop('class', axis=1)    
y_test = tested_set['class']                 


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


k = 6

knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')


knn.fit(X_train_scaled, y_train)


y_pred = knn.predict(X_test_scaled)


accuracy = accuracy_score(y_test, y_pred)


error_rate = 1 - accuracy


conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)


print(f"Value of k: {k}")
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Error Rate: {error_rate * 100:.2f}%")
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```

#### ii. Test Data

```python
def knn_model(k_value):
    knn = KNeighborsClassifier(n_neighbors=k_value, metric='euclidean')
    knn.fit(X_train, y_train)
    return knn


k_values = np.arange(208, 0, -3) 
train_errors = []
test_errors = []


for k in k_values:
    model = knn_model(k)
    
   
    train_pred = model.predict(X_train)
    train_errors.append(1 - accuracy_score(y_train, train_pred))
    
  
    test_pred = model.predict(X_test)
    test_errors.append(1 - accuracy_score(y_test, test_pred))


plt.figure(figsize=(10, 6))
plt.plot(k_values, train_errors, label='Train Error', marker='o')
plt.plot(k_values, test_errors, label='Test Error', marker='o')
plt.xlabel('k value')
plt.ylabel('Error')
plt.title('Train and Test Errors vs k value')
plt.legend()
plt.show()


optimal_k = k_values[np.argmin(test_errors)]
print(f"The most suitable k value is: {optimal_k}")


best_model = knn_model(optimal_k)
y_pred = best_model.predict(X_test)


conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=['Normal', 'Abnormal'])

print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

#### iii. Learning Curve

```python
def find_best_k(X_train_subset, y_train_subset, X_test, y_test, k_values):
    test_errors = []
    
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
        knn.fit(X_train_subset, y_train_subset)
        test_pred = knn.predict(X_test)
        test_errors.append(1 - accuracy_score(y_test, test_pred))
    
    best_k = k_values[np.argmin(test_errors)]
    best_test_error = min(test_errors)
    
    return best_k, best_test_error

N_values = np.arange(10, 211, 10)
best_test_errors = []

for N in N_values:
    
    zero_data_subset = zero_data.iloc[:N//3]
    one_data_subset = one_data.iloc[:N - N//3]
    subset_train_set = pd.concat([zero_data_subset, one_data_subset])
    
   
    X_train_subset = subset_train_set.drop('class', axis=1)
    y_train_subset = subset_train_set['class']
    
   
    k_values = np.arange(1, N, 5)
    
    
    best_k, best_test_error = find_best_k(X_train_subset, y_train_subset, X_test, y_test, k_values)
    
   
    best_test_errors.append(best_test_error)


plt.figure(figsize=(10, 6))
plt.plot(N_values, best_test_errors, label='Best Test Error Rate', marker='o')
plt.xlabel('Training Set Size (N)')
plt.ylabel('Best Test Error Rate')
plt.title('Learning Curve: Best Test Error Rate vs Training Set Size')
plt.legend()
plt.show()
```

### (d) Other Metrics

#### i. Minkowski Distance.

##### A. Manhattan Distance with p = 1.

```python
def knn_with_manhattan(k_value):
    knn = KNeighborsClassifier(n_neighbors=k_value, metric='manhattan')
    knn.fit(X_train, y_train)
    return knn


k_values = np.arange(1, 197, 5)
test_errors = []

for k in k_values:
    model = knn_with_manhattan(k)
    y_pred = model.predict(X_test)
    test_errors.append(1 - accuracy_score(y_test, y_pred))


optimal_k_manhattan = k_values[np.argmin(test_errors)]
print(f"Optimal k for Manhattan distance: {optimal_k_manhattan}")
```

##### B. With log10(p) in {0.1, 0.2, 0.3, ... ,1}.

```python
def knn_with_minkowski(k_value, p_value):
    knn = KNeighborsClassifier(n_neighbors=k_value, metric='minkowski', p=p_value)
    knn.fit(X_train, y_train)
    return knn

log_p_values = np.arange(0.1, 1.1, 0.1)
test_errors_p = []

for log_p in log_p_values:
    p_value = 10**log_p
    model = knn_with_minkowski(optimal_k_manhattan, p_value)
    y_pred = model.predict(X_test)
    test_errors_p.append(1 - accuracy_score(y_test, y_pred))


best_log_p = log_p_values[np.argmin(test_errors_p)]
print(f"Best log10(p) for Minkowski distance: {best_log_p}")
```

##### C. Chebyshev Distance With p -> infinity.

```python
def knn_with_chebyshev(k_value):
    knn = KNeighborsClassifier(n_neighbors=k_value, metric='chebyshev')
    knn.fit(X_train, y_train)
    return knn


test_errors_chebyshev = []

for k in k_values:
    model = knn_with_chebyshev(k)
    y_pred = model.predict(X_test)
    test_errors_chebyshev.append(1 - accuracy_score(y_test, y_pred))


optimal_k_chebyshev = k_values[np.argmin(test_errors_chebyshev)]
print(f"Optimal k for Chebyshev distance: {optimal_k_chebyshev}")
```

#### ii. Mahalanobis Distance.

```python
cov_matrix = np.cov(X_train.T)
inv_cov_matrix = np.linalg.inv(cov_matrix)


def knn_with_mahalanobis(k_value):
    
    mahalanobis_metric = DistanceMetric.get_metric('mahalanobis', VI=inv_cov_matrix)
    
    knn = KNeighborsClassifier(n_neighbors=k_value, metric=mahalanobis_metric)
    knn.fit(X_train, y_train)
    return knn


test_errors_mahalanobis = []

for k in k_values:
    model = knn_with_mahalanobis(k)
    y_pred = model.predict(X_test)
    test_errors_mahalanobis.append(1 - accuracy_score(y_test, y_pred))


optimal_k_mahalanobis = k_values[np.argmin(test_errors_mahalanobis)]
print(f"Optimal k for Mahalanobis distance: {optimal_k_mahalanobis}")
```

```python
summary_data = {
    'Distance Metric': ['Manhattan', 'Minkowski', 'Chebyshev', 'Mahalanobis'],
    'Optimal k': [optimal_k_manhattan, best_log_p, optimal_k_chebyshev, optimal_k_mahalanobis],
    'Best Test Error': [min(test_errors), min(test_errors_p), min(test_errors_chebyshev), min(test_errors_mahalanobis)]
}

summary_table = pd.DataFrame(summary_data)
print(summary_table)
```

### (e) Weighted Decision

```python
def knn_with_weighted_voting(k_value, distance_metric):
    
    knn = KNeighborsClassifier(n_neighbors=k_value, metric=distance_metric, weights='distance')
    knn.fit(X_train, y_train)
    return knn


k_values = np.arange(1, 197, 5)


best_test_errors_weighted = {
    'euclidean': [],
    'manhattan': [],
    'chebyshev': []
}


for distance_metric in ['euclidean', 'manhattan', 'chebyshev']:
    print(f"Testing {distance_metric} distance with weighted voting")
    test_errors = []
    
    for k in k_values:
        model = knn_with_weighted_voting(k, distance_metric)
        y_pred = model.predict(X_test)
        test_errors.append(1 - accuracy_score(y_test, y_pred))
    
 
    best_test_errors_weighted[distance_metric] = test_errors


for distance_metric in best_test_errors_weighted:
    best_error = min(best_test_errors_weighted[distance_metric])
    optimal_k = k_values[np.argmin(best_test_errors_weighted[distance_metric])]
    print(f"Best test error for {distance_metric} distance: {best_error}, with k = {optimal_k}")
```
